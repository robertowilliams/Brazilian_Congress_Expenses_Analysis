### **C. Análise Estatística - Triênio**

#### Será realizada a análise descritiva dos três anos 2015, 2016 e 2017:
> **C.1** Quantidade de gastos por tipo.
>
> **C.2** Montante de gastos por tipo.
>
> **C.3** Montante de gasto por Partido.
>
> **C.4** Montante de gastos por UF (Unidade Federal).
>
> **C.5** Total de Gastos dos Top3 Deputados de maior gasto no triênio 2015, 2016 e 2017.

#### **C.1** Quantidade de gastos por tipo.

```python
# Variáveis auxiliares para plotar gráficos
data = df_limpo
target = 'Tipo'

plt.figure(figsize=(largura,altura))

# Variável com função que ordena as barras em ordem decrescente.
descending_order = data[target].value_counts().sort_values(ascending=False).index
plt.suptitle("Quantidade de Gastos por Tipo", fontsize=fonte)
sns.countplot(data=data,x=target,order=descending_order)
plt.xticks(rotation=rota)
plt.show()
```

#### C.2 Montante de Gastos por Tipo

```python
df_limpo.head(2)
```

```python
# Impedir eixo y de demonstrar valores em notaçào científica.
plt.ticklabel_format(style='plain', axis='y')

# Filtrar dataframe df_limpo por 'Tipo de gasto', valores da coluna 'Valor', realiza seu somatório e ordena de de forma descendente.
df_c2 = df_limpo.groupby('Tipo')['Valor'].sum().sort_values(ascending=False)

# Plota o gráfico em barras.
df_c2.plot( kind='bar', figsize = (largura,altura), rot=rota, sort_columns = True)

plt.suptitle("Montante de Gastos por Tipo", fontsize=fonte)
plt.show()
```

#### C.3 Total de Gastos por Partido

```python
# Impedir eixo y de demonstrar valores em notaçào científica.
plt.ticklabel_format(style='plain', axis='y')

# Filtrar dataframe df_limpo por 'Tipo de gasto', valores da coluna 'Valor', realiza seu somatório e ordena de de forma descendente.
df_c3 = df_limpo.groupby('Partido')['Valor'].sum().sort_values(ascending=False)

# Plota o gráfico em barras.
df_c3.plot( kind='bar', figsize = (largura,altura), rot=rota, sort_columns = True)
plt.suptitle("Montante de Gastos por Partido", fontsize=fonte)
plt.show()
```

#### C.4 Total de Gastos por UF (Unidade Federal)

```python
# Impedir eixo y de demonstrar valores em notaçào científica.
plt.ticklabel_format(style='plain', axis='y')

# Filtrar dataframe df_limpo por 'Tipo de gasto', valores da coluna 'Valor', realiza seu somatório e ordena de de forma descendente.
df_c4 = df_limpo.groupby('UF')['Valor'].sum().sort_values(ascending=False)

# Plota o gráfico em barras.
df_c4.plot( kind='bar', figsize = (largura,altura), rot=rota, sort_columns = True)
plt.suptitle("Total de Gastos por UF (Unidade Federal)", fontsize=fonte)
plt.show()
```

#### C.5 Top 3 Deputados

```python
# Total de Gastos dos Top3 Deputados de maior gasto no triênio 2015, 2016 e 2017.
#rota = 0
df_trio = df_limpo.groupby('Nome')['Valor'].sum().sort_values(ascending=False).head(3)
df_trio.plot(kind='bar', figsize = (largura,altura), rot=rota)
plt.suptitle("Total de Gastos dos Top3 Deputados de maior gasto no trienio 2015, 2016 e 2017", fontsize=fonte)
plt.show()
```

```python
# Definindo dataframe com of top3 deputados
df_top3 = df_limpo[(df_limpo['Nome'] == 'MACEDO') | \
                   (df_limpo['Nome'] == 'FRANCISCO CHAPADINHA') |  \
                   (df_limpo['Nome'] == 'DOMINGOS NETO')]

# Agrupando o dataframe df_top3 pelo feature 'Tipo de gasto', somando a coluna'Valor'
# ordenando os valores de forma descendente, em seguida ploto o gráfico.

df_top3.groupby('Tipo')['Valor'].sum().sort_values(ascending=False).\
plot( kind='bar', figsize = (largura,altura), rot=rota, sort_columns = True)
plt.suptitle("Gastos dos deputado Top 3, por tipo de gasto", fontsize=fonte)

plt.show()
```

```python
# Análise de gastos do Deputado Macedo.
df_macedo = df_limpo[df_limpo['Nome'] == 'MACEDO'].groupby('Tipo').Valor.mean().\
sort_values(ascending=False)
plt.figure(figsize=(largura,altura))

df_macedo.plot(kind='bar')
plt.suptitle("Media de Gastos do Deputado Macedo", fontsize=fonte)
plt.show()
```

```python
# Análise de gastos do Deputado Macedo.
df_macedo = df_limpo[df_limpo['Nome'] == 'FRANCISCO CHAPADINHA'].groupby('Tipo').Valor.mean().\
sort_values(ascending=False)
plt.figure(figsize=(largura,altura))

df_macedo.plot(kind='bar')
plt.suptitle("Media de Gastos do Deputado Francisco Chapadinha", fontsize=fonte)
plt.show()
```

```python
# Análise de gastos do Deputado Macedo.
df_macedo = df_limpo[df_limpo['Nome'] == 'DOMINGOS NETO'].groupby('Tipo').Valor.mean().\
sort_values(ascending=False)
plt.figure(figsize=(largura,altura))

df_macedo.plot(kind='bar')
plt.suptitle("Media de Gastos do Deputado Domingos Neto", fontsize=fonte)
plt.show()
```

### **E. Análise do Dataset**

#### Realizaremos a preparação do dataset para uso de algoritmo de regressão.

```python
# Criação de dataframe com os 3 deputados
#df_dep = df_limpo[df_limpo.Nome.isin(['FRANCISCO CHAPADINHA', 'DOMINGOS NETO' , 'ROCHA'])]

# Criação de dataframe com o dep. Fco Chapadinha
df_fco = df_limpo[(df_limpo['Nome'] == 'FRANCISCO CHAPADINHA')]

# Criação de dataframe com o dep. Domingos Neto
df_dom = df_limpo[(df_limpo['Nome'] == 'DOMINGOS NETO')]

# Criação de dataframe com o dep. Rocha
df_roc = df_limpo[(df_limpo['Nome'] == 'ROCHA')]
```

#### E.2 Dep. Francisco Chapadinha - Somando os tipos de gastos por ano

```python
# Agrupando os gastos por ano
df_fco_2015 = pd.DataFrame(df_fco[df_fco["Ano"] == 2015].groupby('Tipo')['Valor'].sum())
df_fco_2016 = pd.DataFrame(df_fco[df_fco["Ano"] == 2016].groupby('Tipo')['Valor'].sum())
df_fco_2017 = pd.DataFrame(df_fco[df_fco["Ano"] == 2017].groupby('Tipo')['Valor'].sum())

# Criando o dataframe que receberá os dados finais
df_fco_trienio = df_fco_2015
# Renomeando a coluna
df_fco_trienio.rename(columns = {'Valor': 'Valor 2015'}, inplace = True)

# Adicionando o conteúdo de cada dataframe no Dataframe final
df_fco_trienio['Valor 2015'] = df_fco_2015
df_fco_trienio['Valor 2016'] = df_fco_2016['Valor']
df_fco_trienio['Valor 2017'] = df_fco_2017['Valor']

# Exportando para uso na fase de predição
df_fco_trienio.to_csv('./dataset/df_fco_trienio.csv', \
                         index=True, sep=';', encoding = 'latin1') 

df_fco_trienio
```

```python
# Agrupando os gastos por ano
df_fco_2015 = pd.DataFrame(df_fco[df_fco["Ano"] == 2015].groupby('Tipo')['Valor'].sum()).reset_index()
df_fco_2016 = pd.DataFrame(df_fco[df_fco["Ano"] == 2016].groupby('Tipo')['Valor'].sum()).reset_index()
df_fco_2017 = pd.DataFrame(df_fco[df_fco["Ano"] == 2017].groupby('Tipo')['Valor'].sum()).reset_index()

df_fco_2015['Ano'] = 2015
df_fco_2016['Ano'] = 2016
df_fco_2017['Ano'] = 2017

df_fco_con = pd.concat([df_fco_2015, df_fco_2016, df_fco_2017])
```

```python
pd.index_reset(df_fco_con)
```

```python
# Variáveis para definição de dimensões e rotação do gráfico.

df_fco_trienio.plot(kind = 'bar', title='Gastos do Trienio' , fontsize=fonte, \
                    figsize=(largura,altura), subplots = False, legend = True)
plt.show()
```

#### C.3 Dep. Domingos Neto - Somando os tipos de gastos por ano

```python
# Agrupando os gastos por ano
df_dom_2015 = pd.DataFrame(df_dom[df_dom["Ano"] == 2015].groupby('Tipo')['Valor'].sum())
df_dom_2016 = pd.DataFrame(df_dom[df_dom["Ano"] == 2016].groupby('Tipo')['Valor'].sum())
df_dom_2017 = pd.DataFrame(df_dom[df_dom["Ano"] == 2017].groupby('Tipo')['Valor'].sum())

# Criando o dataframe que receberá os dados finais
df_dom_trienio = df_dom_2015
# Renomeando a coluna
df_dom_trienio.rename(columns = {'Valor': 'Valor 2015'}, inplace = True)

# Adicionando o conteúdo de cada dataframe no Dataframe final
df_dom_trienio['Valor 2016'] = df_dom_2016['Valor']
df_dom_trienio['Valor 2017'] = df_dom_2017['Valor']

# Exportando para uso na fase de predição
df_dom_trienio.to_csv('./dataset/df_dom_trienio.csv', \
                         index=True, sep=';', encoding = 'latin1')

df_dom_trienio
```

```python
df_dom_trienio.plot(kind = 'bar', title='Gastos do Trienio', figsize=(largura,altura), fontsize=fonte)
plt.show()
```
